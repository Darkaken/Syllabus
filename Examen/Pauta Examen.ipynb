En una conferencia se realizan varios eventos, tales como charlas, workshops y seminarios, donde distintas personas exponen sobre un tema. El principal desafío es organizar el orden de estos eventos. En cada horario pueden haber varios eventos simultáneamente; no todas las salas tienen la misma capacidad y no todos los temas son igual de populares. Usted debe diseñar un programa que evalúe planificaciones. Cada planificación viene en un archivo con formato `JSON` e indica cuáles son los asistentes, las salas, los eventos y los horarios. 

Cada evento tiene un nombre, horario y una sala asignada. Cada sala tiene un nombre y capacidad máxima. Para asistir a un evento, los asistentes deben hacer una cola afuera de la sala correspondiente. La felicidad de un asistente es la suma de la felicidad que le entregan los eventos a los que pudo entrar. Cada asistente le asigna a cada evento un puntaje equivalente a la felicidad que el evento le proporciona. El puntaje va de 0 a 10. La felicidad total del evento se calcula como:

$$

$$

donde $ x_{i, e} $ vale 1 en el caso de que el asistente $i$ entre al evento $e$, y -1 en caso contrario, e $ y_{i, e} $ vale 1 si el asistente espera en la cola para entrar al evento y 0 en caso contrario.

### Solución

**(5 ptos):  Manejo del tiempo**
Clase Simulación debe manejar en una variable el tiempo actual de la simulación. En el método `run()` (o equivalente) de la clase Sim. se deben inicializar threads para cada sala (o sus colas correspondientes). Esto permitirá que las colas avancen paralelamente y es necesario para lograr que estas se sincronizen. 

También es válido mencionar que cada asistente será un thread, pero las colas también deben serlo. En caso de que no se mencione 3 puntos.

**(3 pts)**:
Dado que la simulación es de tipo DES, cada sala debe tener una lista de eventos a ocurrir con su tiempo correspondiente. Se calcula el tiempo mínimo de estos eventos. Cada thread tendrá un tiempo_prox_evento con este tiempo min. Al terminar un evento, este tiempo_prox_evento se actualiza con el nuevo mínimo. Cuando se términa un evento, sin importar de que thread, se calcula el mínimo de los tiempo_prox_evento de todas las salas (colas). Se actualiza el tiempo de la simulación. Para que cada thread este sincronizado, se propone que sea de la forma:
```
while True:
    if tiempo_simulacion == tiempo_prox_evento:
        # realizar evento correspondiente
```
**(5 ptos): Threads y Locks**
Cuando se cambia el tiempo de la simulación, varios threads pueden realizar un evento de manera paralela. Es necesario hacer un lock sobre la cola de la sala al realizar un evento, para evitar que, por ejemplo, una persona se cambie de cola y entre a la otra sala en el mismo tiempo de simulación, lo que no tiene sentido. Se debe hacer un lock sobre la cola, dejar entrar los asistentes, terminar el evento y, finalmente, sumar personas a la fila.

**(25 pts.)** Considere la siguiente estructura de datos para un grafo **dirigido** que almacena ciudades (nodos) y rutas directas (aristas dirigidas) entre ciudades.

````python
class Node:
    
    def __init__(self, ciudad):
        self.ciudad = ciudad
        self.destinos = []
        
    def add_connection(self, node):
        self.destinos.append(node)
        
    def __repr__(self):
        return self.ciudad + " -> (" + ",".join([c.ciudad for c in self.destinos]) + ")"
````

````python
# Datos para testear
cities = ["ANF","SCL","PMC","PUQ","ARI","CJC","IQQ","CCP","BBA"]
nodes = {}

for c in cities:
    n = Node(c)
    nodes[c] = n
    
nodes["SCL"].add_connection(nodes["ANF"])
nodes["SCL"].add_connection(nodes["CCP"])
nodes["SCL"].add_connection(nodes["ARI"])
nodes["SCL"].add_connection(nodes["BBA"])

nodes["ANF"].add_connection(nodes["ARI"])
nodes["ANF"].add_connection(nodes["IQQ"])
nodes["ANF"].add_connection(nodes["CJC"])
nodes["ANF"].add_connection(nodes["SCL"])

nodes["ARI"].add_connection(nodes["IQQ"])
nodes["ARI"].add_connection(nodes["SCL"])
nodes["ARI"].add_connection(nodes["CJC"])

nodes["CJC"].add_connection(nodes["SCL"])
nodes["CJC"].add_connection(nodes["ANF"])

nodes["IQQ"].add_connection(nodes["ARI"])

nodes["CCP"].add_connection(nodes["PMC"])
nodes["CCP"].add_connection(nodes["SCL"])

nodes["PMC"].add_connection(nodes["BBA"])
nodes["PMC"].add_connection(nodes["PUQ"])

nodes["PUQ"].add_connection(nodes["SCL"])

nodes["BBA"].add_connection(nodes["CCP"])
````

* **(9 pts.)** Escriba la función `recorrer(n)` que reciba como parámetro un objeto `n` de clase `Node` y, a partir de él, retorne una lista con todas las ciudades alcanzables a partir de `n`, sin repetirlas.

````python
from collections import deque

def recorrer(n):
    q = deque()
    q.append(n)
    visited = []
    while len(q) > 0:
        current = q.pop()
        if current not in visited:
            print("Visiting "+ str(current))
            visited.append(current)
            for c in current.destinos:
                q.append(c)
    return visited
````
